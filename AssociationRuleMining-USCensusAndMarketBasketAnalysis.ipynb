```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

pd.set_option("max_colwidth", 150)
```

```python
f = "https://github.com/cs6220/cs6220.spring2019/raw/master/data/Online%20Retail.xlsx"
df = pd.read_excel(f)
df.head()
```

```python
basket = (df[df['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
basket_sets = basket.applymap(lambda x: 1 if x >= 1 else 0)
```

```python
basket_sets.head()
```

```python
frequent_one_itemsets = apriori(basket_sets, min_support=0.062, use_colnames=True)
```

```python
frequent_one_itemsets 
```

```python
frequent_two_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)
```

```python
frequent_two_itemsets = frequent_two_itemsets[236:241]
```

```python
frequent_two_itemsets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.025, use_colnames=True, max_len=2)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold= 0.2)
```

```python
rules.head()
```

```python
import numpy as np
import pandas as pd
path ="https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/adult/"
names = pd.read_csv(path +"adult.names", sep="\n", header=None)
parse_cols = lambda x: x.str.split(":", expand=True).iloc[:, 0]
columns = np.roll(parse_cols(names.iloc[92:108, 0]), shift=-1)
df_adult = pd.read_csv(path +"adult.data", sep=",", header=None, index_col=False)
df_adult.columns = columns
```

```python
df_adult
```

```python
df_adult = df_adult.drop('age', axis=1)
df_adult = df_adult.drop('fnlwgt', axis=1)
df_adult = df_adult.drop('education-num', axis=1)
df_adult = df_adult.drop('capital-gain', axis=1)
df_adult = df_adult.drop('capital-loss', axis=1)
df_adult = df_adult.drop('hours-per-week', axis=1)
```

```python
df_adult
```

```python
df = pd.get_dummies(df_adult)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True, max_len=3)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold= 0.8)
```
